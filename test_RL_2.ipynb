```
class DQNAgent:
    
    def __init__(self, env, 
                 gamma = .99, 
                 lr_rate = 0.0005, 
                 eps_begin = 1.0, 
                 eps_end = 0.01, 
                 eps_decay = 0.995, 
                 batch_size = 64,
                 replay_capacity = int(1e6),
                 tau = 1e-3, 
                 no_of_neurons = 64):
        
        self.env = env
        self.gamma = gamma
        self.lr_rate = lr_rate
        self.eps = eps_begin
        self.eps_end = eps_end
        self.eps_decay = eps_decay
        self.eps_hist = []
        self.batch_size = batch_size
        self.replay_capacity = replay_capacity
        self.tau = tau
        self.opt = Adam(lr = self.lr_rate)
        self.no_of_neurons = no_of_neurons
        self.online_model = self.construct_model()
        self.target_model = self.construct_model()
        self.memory  = deque(maxlen=int(1e5))
        self.step = 0
        self.step_per_lr = 4
        self.episode_reward = 0
        self.episode_size = 0
        self.episode = 0
        self.reward_hist = deque(maxlen=100)
        self.episodes_steps = []
    
    def construct_model(self):
        model = Sequential()
        model.add(Dense(self.no_of_neurons, input_dim= self.env.observation_space.shape[0], activation="relu"))
        model.add(Dense(self.no_of_neurons, activation="relu"))
        model.add(Dense(self.env.action_space.n))
        model.compile(loss="mean_squared_error",optimizer= self.opt)
        return model
    
    def update_eps(self):
        self.eps*= self.eps_decay
        self.eps = max(self.eps, self.eps_end)
    
    def choose_action(self, state):
        if np.random.random() > self.eps:
            state_reshape = np.reshape(state, [1, self.env.observation_space.shape[0]])
            q_values = self.online_model.predict(state_reshape)
            action = np.argmax(q_values[0])
        else:
            action = self.env.action_space.sample()
        return action
    
    def store_memory(self, state, action, reward, next_state, done):
        self.memory.append([state, action, reward, next_state, done])
        self.step += 1
        self.episode_reward += reward
        self.episode_size += 1
        #print("------Episode size {} / steps {}-------".format(self.episode_size,self.step))
        if done:
            self.eps_hist.append(self.eps)
            self.update_eps()
            self.episode += 1
            self.reward_hist.append(self.episode_reward)
            self.episodes_steps.append(self.episode_size)
            self.episode_reward = 0
            self.episode_size = 0

    def target_model_learning(self): 
        online_weight = self.online_model.get_weights()
        target_weight = self.target_model.get_weights()
        for i in range(len(target_weight)):
            target_weight[i] = (1-self.tau)*target_weight[i] + self.tau*online_weight[i]
        self.target_model.set_weights(target_weight)
            
    def online_model_learning(self):
        if self.step % self.step_per_lr != 0:
            return
        
        if self.batch_size > len(self.memory):
            return
        
        state_ls = []
        target_fct_ls = []
        
        learning_batch = random.sample(self.memory, self.batch_size)
        
        #features = np.zeros((self.batch_size,self.env.observation_space.shape[0]))
        #labels = np.zeros((self.batch_size,self.env.action_space.n))
        
        count = 0
        for sample in learning_batch:
            state = np.reshape(sample[0], [1, self.env.observation_space.shape[0]]) 
            action = sample[1]
            reward = sample[2]
            next_state = np.reshape(sample[3], [1, self.env.observation_space.shape[0]]) 
            done = sample[4]
            #print("---------------next_state--------------")
            #print(next_state)
            #print("---------------self.target_model.predict(next_state)--------------")
            #print(self.target_model.predict(next_state))
            #print("---------------self.target_model.predict(next_state)[0]--------------")
            #print(self.target_model.predict(next_state)[0])
            #print("---------------np.amax(self.target_model.predict(next_state)[0])--------------")
            #print(np.amax(self.target_model.predict(next_state)[0]))            
            
            if done:
                td_target = reward
            else:
                max_q_value = np.amax(self.target_model.predict(next_state)[0])
                td_target = reward + self.gamma * max_q_value
            
            current_q_values = self.online_model.predict(state)
            current_q_values[0][action] = td_target
            
            state_ls.append(state[0])
            target_fct_ls.append(current_q_values[0])
            
            count +=1
        self.online_model.fit(np.array(state_ls),np.array(target_fct_ls),epochs=1,verbose=0)
        self.target_model_learning()
    
    def store_output_as_csv(self,name):
        output = pd.DataFrame({'Reward': self.reward_hist, "eps": self.eps_hist, "episodes_steps": self.episodes_steps})
        output.to_csv('{}.csv'.format(name))
        return output
        
    def save_model(self,name):
        self.online_model.save("{}_online.h5".format(name))
        self.target_model.save("{}_target.h5".format(name))
```

```
def train_model(gamma, lr_rate, eps, eps_decay, max_episodes, max_steps, csv_name,model_name):
    env = gym.make("LunarLander-v2")
    agent = DQNAgent(env = env, gamma = gamma, lr_rate = lr_rate, eps_begin = eps, eps_decay = eps_decay)
    episode = 0
    reward_ls = []
    epsilon_ls = []
    for epis in range(max_episodes):
        #print("-------------------Episode {}-------------------".format(episode))
        current_state = env.reset()
        reward_eps = 0
        for i in range(max_steps):
            action = agent.choose_action(current_state)
            next_state, reward, done, _ = env.step(action)
            reward_eps += reward
            agent.store_memory(current_state, action, reward, next_state, done)
            agent.online_model_learning()
            current_state = next_state
            if done:
                break
        reward_ls.append(reward_eps)
        epsilon_ls.append(agent.eps_hist[-1])
        if np.mean(reward_ls[-100:]) > 200:
            break
        episode += 1
        latest_reward = reward_ls[-1]
        latest_eps = agent.eps_hist[-1]
        print("-------------------Episode {} : Reward {} / MA_Reward {} / Epsilon {}-------------------".format(episode, latest_reward, np.mean(reward_ls[-100:]), latest_eps))
        #if agent.episode % 100 == 0:
        #    agent_episode = str(agent.episode)
        #    output = agent.store_output_as_csv(csv_name + "_" + agent_episode)
    agent.save_model(model_name)
    env.close()
    return reward_ls, epsilon_ls
```

```
def test_model(model):
    test_model = load_model(model)
    culmulated_reward_ls = []
    env = gym.make("LunarLander-v2")
    episode = 0
    for i in range(100):
        culmulated_reward = 0
        current_state = np.reshape(env.reset(), [1, env.observation_space.shape[0]])
        done = False
        while not done:
            action = np.argmax(test_model.predict(current_state)[0])
            new_state, reward, done, _ = env.step(action)
            current_state = np.reshape(new_state , [1, env.observation_space.shape[0]])
            culmulated_reward += reward
        culmulated_reward_ls.append(culmulated_reward)
        episode += 1
        print("-------------------Episode {} : Reward {}-------------------".format(episode,culmulated_reward))
    env.close()
    return culmulated_reward_ls
```
